위는 XML 마크업 언어로 되어있다. 필요한 정보만 가져오기 위해서는 몇몇 정보를 파싱해야 한다.

company() #기업의 정보를 불러옴 #인자로 기업의 종목 코드, 고유번호, 회사이름 중 하나를 받는다.

```python
print(dart.company("하나금융지주"))
```

report() #사업보고서의 주요 정보를 가져온다. #회사이름, key_word, 사업연도, 보고서 코드를 인자로 받는다

```python
display(dart.report("하나금융지주", "배당", 2023, "11011")) #11011은 사업보고서 코드
```

finstate() #기업의 재무 제표를 가져온다. #report와 똑같은 인자를 가진다.

```python
display(dart.finstate("하나금융지주", 2023, "11011"))
```

```python
hana = dart.finstate("하나금융지주", 2023, "11011")
hana = hana.loc[hana['fs_nm']=="연결재무제표"] #fs_nm: 연결명 / sj_nm: 재무제표 명
hana = hana.loc[hana["account_nm"].isin(["당기순이익","영업이익","매출액"])] # account_nm: 계정명/ 그 중 당기순이익, 영업이익, 매출액을 추출하는 명령어
hana = hana.loc[:,["account_nm","fs_nm","sj_nm","thstrm_dt","thstrm_amount"]] # 모든 행에서 다음과 같은 컬럼을 추출해라
display(hana)
```
